```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_mask(mask, ax, random_color=False):
    if random_color:
        color = np.concatenate([np.random.random(3), np.array([0.6])], axis=0)
    else:
        color = np.array([30 / 255, 144 / 255, 255 / 255, 0.4])
    h, w = mask.shape[-2:]
    mask_image = mask.reshape(h, w, 1) * color.reshape(1, 1, -1)
    ax.imshow(mask_image)
```

```python
data_folder_path = Path('data')
image_name = 'MCUCXR_0001_0'
image_file_path = Path(f'image_{image_name}.npz')
masks_file_path = Path(f'masks_{image_name}.npz')
image_slice_path = data_folder_path / image_file_path
masks_slice_path = data_folder_path / masks_file_path
```

```python
image_slice_npz = np.load(image_slice_path)
image_slice_npz_keys = list(image_slice_npz.keys())
image_slice = image_slice_npz[image_slice_npz_keys[0]][:, :, 0]
image_slice.shape
```

```python
masks_slice_npz = np.load(masks_slice_path)
masks_slice_npz_keys = list(masks_slice_npz.keys())
masks_slice = masks_slice_npz[masks_slice_npz_keys[0]][:, :, 0]
masks_slice.shape
```

```python
plt.figure(figsize=(10, 10))
plt.style.use('grayscale')
plt.imshow(image_slice)
show_mask(masks_slice, plt.gca())
```
