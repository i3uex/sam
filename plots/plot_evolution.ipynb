```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

df1 = pd.read_csv('../working_data/results.backup/2023-05-19/2. Todo el dataset/1/image_coronacases_001/raw_data_2023-05-19_15-47-40.csv')
df2 = pd.read_csv('../working_data/results.backup/2023-05-19/2. Todo el dataset/2/image_coronacases_001/raw_data_2023-05-19_16-50-56.csv')
df3 = pd.read_csv('../working_data/results.backup/2023-05-19/2. Todo el dataset/3/image_coronacases_001/raw_data_2023-05-19_18-19-58.csv')
df4 = pd.read_csv('../working_data/results.backup/2023-05-19/2. Todo el dataset/4/image_coronacases_001/raw_data_2023-05-19_19-20-54.csv')

ax = plt.gca()

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))

ax.set_xlim(100, 250)
ax.set_ylim(0.5, 1.0)

labels = [
    'Baseline (62.93 %)',
    '+ centroids inside (78.39%)',
    '+ negative prompt (91.26%)',
    '+ bounding box (92.79%)'
]

df1.plot(kind='line', x='slice', y='jaccard', ax=ax)
df2.plot(kind='line', x='slice', y='jaccard', ax=ax)
df3.plot(kind='line', x='slice', y='jaccard', ax=ax)
df4.plot(kind='line', x='slice', y='jaccard', ax=ax)

plt.xlabel('Slice')
plt.ylabel('Jaccard index')
plt.legend(labels)
plt.show()
```
